La valeur aberrante correspond à l’année 2020 aux États-Unis

```python
data = df[(df['Country']=='USA') &
          (df['Year']>=2000) &
          (df['Year']<=2020)]

plt.figure(figsize=(12,8))
sns.scatterplot(data=data, x='Spending_USD', y="Life_Expectancy")

# Ajouter les années sur chaque point
for i in range(len(data)):
    plt.text(data['Spending_USD'].iloc[i],
             data['Life_Expectancy'].iloc[i],
             str(data['Year'].iloc[i]),
             fontsize=9)

plt.xlabel("Spending USD")
plt.ylabel("Life Expectancy")
plt.title("USA: Spending vs Life Expectancy (2000-2020)")
plt.show()
```
